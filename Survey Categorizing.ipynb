```python
categories = pd.read_excel("H20 Customer Survey 2020-2021 Comments.xlsx", sheet_name="Issue Categories")
survey_comments = pd.read_excel("H20 Customer Survey 2020-2021 Comments.xlsx", sheet_name="Improve Experience")
```

```python
counter = 0
category_counter = 0
category_master_list = []

access_list = []
locals_list = []
permits_list = []
alt_trans_list = []
dogs_list =  []
congestion_list = []
cost_list = []
equity_list = []
infrastructure_list = []
litter_list = []
management_list = []
modal_list =[]
natural_list = []
parking_list = []
reservation_list = []
safety_list = []
technology_list = []
experience_list = []
education_list = []
wayfinding_list = []
winter_list = []


for x in survey_comments['What could we do to improve the experience of your visit? ']:
    
    category_list = []
    access = "No"
    local= "No"
    permits= "No"
    alt_trans= "No"
    dogs= "No"
    congestion= "No"
    cost= "No"
    equity= "No"
    infrastructure= "No"
    litter= "No"
    management= "No"
    modal= "No"
    natural= "No"
    parking= "No"
    reservation= "No"
    safety= "No"
    technology= "No"
    experience= "No"
    education= "No"
    wayfinding= "No"
    winter= "No"
    for keywords in categories['Keywords']:
        keyword_list = keywords.split(";")

        for keyword in keyword_list:
            #print(keyword)
            if keyword in str(x):
                if category_counter == 0:
                    access = "Yes"
                if category_counter == 1:
                    local = "Yes"
                if category_counter == 2:
                    permits = "Yes"
                if category_counter == 3:
                    alt_trans = "Yes"
                if category_counter == 4:
                    dogs = "Yes"
                if category_counter == 5:
                    congestion = "Yes"
                if category_counter == 6:
                    cost = "Yes"
                if category_counter == 7:
                    equity = "Yes"
                if category_counter ==8:
                    infrastructure = "Yes"
                if category_counter == 9:
                    litter = "Yes"
                if category_counter == 10:
                    management = "Yes"
                if category_counter == 11:
                    modal = "Yes"
                if category_counter == 12:
                    natural = "Yes"
                if category_counter == 13:
                    parking = "Yes"
                if category_counter == 14:
                    reservation = "Yes"
                if category_counter == 15:
                    parking = "Yes"
                if category_counter == 16:
                    safety = "Yes"
                if category_counter == 17:
                    technology = "Yes"
                if category_counter == 18:
                    experience = "Yes"
                if category_counter == 19:
                    education = "Yes"
                if category_counter == 20:
                    wayfinding = "Yes"
                if category_counter == 21:
                    winter = "Yes"
                
                category_list.append(categories['Category'][category_counter])
                break
       # print(categories['Category'][category_counter])
        category_counter +=1
        
        
    counter +=1
    category_counter = 0

    access_list.append(access)
    locals_list.append(local)
    permits_list.append(permits)
    alt_trans_list.append(alt_trans)
    dogs_list.append(dogs)
    congestion_list.append(congestion)
    cost_list.append(cost)
    equity_list.append(equity)
    infrastructure_list.append(infrastructure)
    litter_list.append(litter)
    management_list.append(management)
    modal_list.append(modal)
    natural_list.append(natural)
    parking_list.append(parking)
    reservation_list.append(reservation)
    safety_list.append(safety)
    technology_list.append(technology)
    experience_list.append(experience)
    education_list.append(education)
    wayfinding_list.append(wayfinding)
    winter_list.append(winter)
```

```python
survey_comments['Access']= access_list
survey_comments['Access for Locals']= locals_list
survey_comments['Activity Permits']= permits_list
survey_comments['Alternative Transportation']= alt_trans_list
survey_comments['Concerns with Dogs']= dogs_list
survey_comments['Congestion']= congestion_list
survey_comments['Cost']= cost_list
survey_comments['Equity']= equity_list
survey_comments['Infrastructure and Facilities']= infrastructure_list
survey_comments['Litter/Waste']= litter_list
survey_comments['Management and Coordination']= management_list
survey_comments['Modal Conflicts']= modal_list
survey_comments['Natural Resources and Scenic Quality']= natural_list
survey_comments['Parking']= parking_list
survey_comments['Reservation System']= reservation_list
survey_comments['Safety']= safety_list
survey_comments['Technology']= technology_list
survey_comments['User Experience']= experience_list
survey_comments['Visitor Education']= education_list
survey_comments['Wayfinding ']= wayfinding_list
survey_comments['Winter Activites']= winter_list
```

```python
fields = ['Access', 'Access for Locals', 'Activity Permits', 'Alternative Transportation', 'Concerns with Dogs', 'Congestion', 'Cost', 'Equity', 
          'Infrastructure and Facilities', 'Litter/Waste', 'Management and Coordination', 'Modal Conflicts', 'Natural Resources and Scenic Quality', 
          'Parking', 'Reservation System', 'Safety', 'Technology', 'User Experience', 'Visitor Education', 'Wayfinding ', 'Winter Activites']
```

```python
num_yes_list = []

for field in fields:
    num_yes = len(survey_comments.loc[survey_comments[field]=="Yes"])
    print(field, ": ", num_yes, ", ", round(num_yes/len(survey_comments)*100,2), "% of survey respondents")
    num_yes_list.append(num_yes)
                  
```

```python
import matplotlib.ticker as ticker
plt.style.use('seaborn')
sns.set(rc={'axes.facecolor':'1', 'figure.facecolor':'white', 'patch.linewidth': 0.0})


bar_labels = fields

fig, ax = plt.subplots(1,1,figsize=(20,15))
#ax.set_xticks(fontsize=26, rotation = 90)
#ax.set_yticks(fontsize=26)
ax.set_title('Maroon Bells Survey Comments 2020-2021',fontsize= 34) # title of plot
bars = ax.barh(width = num_yes_list, y = bar_labels)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Number of Survey Respondents Mentioning This Category',fontsize = 30) #xlabel
ax.set_ylabel('Maroon Bells Survey Category', fontsize = 30)#ylabel
ax.tick_params(axis='x', which='major', labelsize=26)
ax.tick_params(axis='y', which='major', labelsize=26)
for bar in bars:
    width = bar.get_width() #Previously we got the height
    label = '{:,}'.format(width)
    label_y_pos = bar.get_y() + bar.get_height() / 2
    ax.text(width*1.02, label_y_pos, s='{:,}'.format(width), va='center',fontsize=26)

for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(18)

plt.savefig("Maroon Bells comments figure.png", bbox_inches='tight')
plt.show()
```
